```python
#CSDL
ds_then = [
    #(d_min, d_max, bề rộng, chiều cao, sâu trục, sâu bạc)
    #0          1       2       3           4       5
    (6,         8,      2,      2,          1.2,    1),
    (8,         10,     3,      3,          1.8,    1.4),
    (10,        12,     4,      4,          2.5,    1.8)
]
vat_lieu_then = [
    ("C45",360),
    ("40Cr",560)
]
#Tạo hàm tính then
import math
def tinh_torque (cong_suat, toc_do):
    #Tính toán torque truyền qua then
    #Khối lệnh
    torque = cong_suat*30000/(math.pi*toc_do)
    return torque

def ung_suat_cho_phep (us_chay=0, hs_an_toan = 3.5, nhiet_luyen=1):
    us_dap_cp = us_chay/(hs_an_toan*nhiet_luyen)
    us_cat_cp = us_dap_cp/2
    return (us_dap_cp,us_cat_cp)

def us_tren_then (i_torque,i_duong_kinh,i_chieu_dai_then,i_be_rong_then,i_chieu_cao_then,i_sau_truc):
    us_dap_tt = 2*i_torque/(i_duong_kinh*i_chieu_dai_then*(i_chieu_cao_then-i_sau_truc))
    us_cat_tt = 2*i_torque/(i_duong_kinh*i_chieu_dai_then*i_be_rong_then)
    return (us_dap_tt,us_cat_tt)

#Tính toán mẫu, mô phỏng các giá trị nhập vào
#chưa có loại trừ trường hợp ngoại lệ

#input
duong_kinh = 8
chieu_dai_truc = 20
cong_suat_truyen = 5
toc_do_quay = 10
vat_lieu_lam_then = "c45"

#tính toán
chieu_dai_then = 0.8*chieu_dai_truc
torque_truyen = tinh_torque(cong_suat_truyen,toc_do_quay)
#lấy kích thước then
for i in range(len(ds_then)):
    if duong_kinh > ds_then[i][0] and duong_kinh <= ds_then[i][1]:
        print(f"thông tin then {ds_then[i]}")
        be_rong_then = ds_then[i][2]
        chieu_cao_then = ds_then[i][3]
        chieu_sau_truc = ds_then[i][4]
        chieu_sau_bac = ds_then[i][5]
for k in range(len(vat_lieu_then)):
    if vat_lieu_lam_then == vat_lieu_then[k][0]:
        us_then = vat_lieu_then[k][1]
print(f"us_then {us_then}")
us_cp = ung_suat_cho_phep(us_then)
us_dap_cp = us_cp[0]
print(us_dap_cp)
us_then = us_tren_then(torque_truyen,duong_kinh,chieu_dai_then,be_rong_then,chieu_cao_then,chieu_sau_truc)
print(us_then)
he_so_sf_dap = us_dap_cp/us_then[0]

if us_then[0] < us_dap_cp[0]:
    print(f"Ứng suất dập trên then là {us_then[0]} MPa")
    print("then du ben dap")
else:
    print("then khong du ben")
print(f"Hệ số an toàn dập là {he_so_sf_dap}")
if he_so_sf_dap > 1.5:
    print(f"hệ số an toàn dập đạt yêu cầu")
else:
    print("hệ số an toàn dập không đạt yêu cầu")
```

```python
import json

# Load dữ liệu JSON một lần
with open("data/materials.json", "r", encoding="utf-8") as f:
    materials = json.load(f)

with open("data/then_standard.json", "r", encoding="utf-8") as f:
    then_data = json.load(f)


def tinh_toan_key(d, l, p, n, eta, z, mat1, mat2):
    """
    Hàm tính toán then
    d: đường kính trục (mm)
    l: chiều dài trục (mm)
    p: công suất (kW)
    n: tốc độ quay (vòng/phút)
    eta: hiệu suất truyền
    z: số then
    mat1, mat2: tên vật liệu
    """
    
    #---Lấy data----
    # Tra kích thước then trong bảng then_data
    b, h, t1, t2 = None, None, None, None
    for item in then_data:
        if item["d_min"] < d <= item["d_max"]:
            b, h, t1, t2 = item["b"], item["h"], item["t1"], item["t2"]
            break

    if b is None or h is None:
        return {"Error": f"Đường kính {d} mm nằm ngoài phạm vi tính toán trong bảng then."}

    # Tra σ của vật liệu
    sigma1 = next((m["sigma"] for m in materials if m["name"] == mat1), None)
    sigma2 = next((m["sigma"] for m in materials if m["name"] == mat2), None)

    #---Tính toán----
    # Moment xoắn (Nm)
    try:
        T = 30000 * p * eta / (3.1416 *n)
    except ZeroDivisionError:
        return {"error": "Tốc độ quay lớn hơn 0"}
    
    # Chiều dài then
    l_then = 0.8 * l

    # Tính ứng suất trên then
    sigma_dap = 2*T/(d*l*(h-t1))
    tau = 2*T/(d*l*b)

    # Tính ứng suất cho phép
    sigma1_cp = sigma1 / 3.5
    tau1_cp = sigma1_cp / 2
    sigma2_cp = sigma2 / 3.5
    tau2_cp = sigma2_cp / 2

    #Tính sf theo vật liệu
    sf11 = sigma1_cp / sigma_dap
    sf12 = tau1_cp / tau
    sf21 = sigma2_cp / sigma_dap
    sf22 = tau2_cp / tau

    #---Đánh giá thiết kế ---
    if 1.5 <= sf11 <= 2 and 1.5 <= sf12 <=2:
        danh_gia1 = str(f"Thiết kế then và sử dụng vật liệu {mat1} đã hợp lý")
    elif sf11 < 1.5 or sf12 < 1.5:
        danh_gia1 = str(f"Thiết kế then và sử dụng vật liệu {mat1} không đủ bền")
    elif sf11 > 2 and sf12 > 2:
        danh_gia1 = str(f"Thiết kế then và sử dụng vật liệu {mat1} thừa bền")
    else:
        danh_gia1 = str(f"Thiết kế then và sử dụng vật liệu {mat1} chưa tối ưu")
    if 1.5 <= sf21 <= 2 and 1.5 <= sf22 <=2:
        danh_gia2 = str(f"Thiết kế then và sử dụng vật liệu {mat2} đã hợp lý")
    elif sf21 < 1.5 or sf22 < 1.5:
        danh_gia2 = str(f"Thiết kế then và sử dụng vật liệu {mat2} không đủ bền")
    elif sf21 > 2 and sf22 > 2:
        danh_gia2 = str(f"Thiết kế then và sử dụng vật liệu {mat2} thừa bền")
    else:
        danh_gia2 = str(f"Thiết kế then và sử dụng vật liệu {mat2} chưa tối ưu")
    # Trả kết quả
    return {
        # Input
        "d": d,
        "l": l,
        "p": p,
        "n": n,
        "eta": eta,
        "z": z,
        "mat1": mat1,
        "mat2": mat2,
        # Output
        "T": round(T, 2),
        "b": b,
        "h": h,
        "t1": t1,
        "t2": t2,
        "l_then": l_then,
        "sigma_dap": sigma_dap,
        "tau": tau,
        "sigma1": sigma1,
        "sigma2": sigma2,
        "sf11": sf11,
        "sf12": sf12,
        "sf21": sf21,
        "sf22": sf22,
        "danh_gia1": danh_gia1,
        "danh_gia2": danh_gia2
    }
demo = tinh_toan_key(25,20,1,20,90,1,"C45","40Cr")

for k, v in demo.items():
    print(f"{k} : {v}")
```

```python
import matplotlib.pyplot as plt

def ve_bieu_do(d, stress_mat1, stress_mat2, stress_limit):
    # Tạo danh sách 5 giá trị đường kính
    d_values = [d-5, d-2.5, d, d+2.5, d+5]

    # Ví dụ: tính ứng suất theo hàm hoặc dữ liệu từ tính toán
    # Ở đây giả sử stress_mat1, stress_mat2 là list ứng suất tương ứng với 5 giá trị đường kính
    # Nếu bạn chỉ có công thức => thay thế đoạn này để tính
    y_mat1 = stress_mat1  
    y_mat2 = stress_mat2  

    plt.figure(figsize=(8,6))

    # Vẽ đường ứng suất cho mat1, mat2
    plt.plot(d_values, y_mat1, marker="o", label="Ứng suất vật liệu 1")
    plt.plot(d_values, y_mat2, marker="s", label="Ứng suất vật liệu 2")

    # Vẽ đường ngang giới hạn
    plt.axhline(y=stress_limit, color="r", linestyle="--", label="Ứng suất khuyến cáo")

    # Thêm tiêu đề, nhãn trục
    plt.title("Biểu đồ ứng suất theo đường kính then")
    plt.xlabel("Đường kính trục (mm)")
    plt.ylabel("Ứng suất (MPa)")

    # Hiển thị chú thích
    plt.legend()

    # Hiển thị grid cho dễ nhìn
    plt.grid(True, linestyle="--", alpha=0.6)

    plt.show()


# Ví dụ test
d = 30
stress_mat1 = [120, 140, 160, 155, 150]   # ứng suất của mat1
stress_mat2 = [100, 110, 130, 135, 128]   # ứng suất của mat2
stress_limit = 150  # giới hạn khuyến cáo

ve_bieu_do(d, stress_mat1, stress_mat2, stress_limit)
```
